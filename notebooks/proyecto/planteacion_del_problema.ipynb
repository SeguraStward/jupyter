# Problem solving shining like a Diamond

- Primero ocupamos limpiar el texto
  - Tokenizacion: dividir el texto en palabras individuales
  - stopWords: eliminar palabras que no aportan nipistola
  - Stemming/lemmatization: reducimos las palabras a sus raices ( gano, ganando pasan a ser gan)

- Vectorizacion(convertir palabras en numeros)
  - el metodo mas comun para hacer esto es Term-frequency-inverse document frecuency)
 

- Modelos de clasificacion:
    - **naive bayes:** se pregunta la probabilidad de que un mensaje sea spam si contiene cierta palabra
    - **Logistic regression:** separa los datos entre spam y no spam
