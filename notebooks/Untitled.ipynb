```python
import heapq

def a_estrella_tramposo():
    # Nuestro mapa con los costos reales de moverse
    grafo = {
        'S': {'A': 3, 'B': 1},
        'B': {'A': 1},         # El atajo secreto de B hacia A
        'A': {'G': 2},
        'G': {}
    }
    
    # Nuestra Heurística (Admisible, pero Inconsistente en A)
    h = {'S': 4, 'B': 3, 'A': 0, 'G': 0}
    
    # Cola de prioridad para explorar (guarda tuplas: (f_costo, nodo))
    frontera = [(h['S'], 'S')]
    g_costos = {'S': 0}
    cerrados = set() # Aquí metemos los nodos que "ya terminamos"
    
    print("--- INICIANDO A* ---")
    
    while frontera:
        f_actual, actual = heapq.heappop(frontera)
        
        # Si sacamos un nodo que ya cerramos y no lo borramos de la cola, lo saltamos
        if actual in cerrados: continue
            
        if actual == 'G':
            print(f"\n🎯 ¡META ALCANZADA! Costo óptimo final: {g_costos['G']}")
            break
            
        # Al sacar el nodo, A* jura que ya encontró el camino más corto hacia él
        cerrados.add(actual)
        print(f"\n🔒 CERRAMOS el nodo '{actual}'. A* cree que ya no hay mejor ruta hacia él.")
        
        # Revisamos vecinos
        for vecino, costo_paso in grafo[actual].items():
            nuevo_g = g_costos[actual] + costo_paso
            
            # Si es la primera vez que vemos al vecino, o encontramos una ruta MÁS BARATA
            if vecino not in g_costos or nuevo_g < g_costos[vecino]:
                
                # ¡AQUÍ ESTÁ LA TRAMPA DE LA INCONSISTENCIA!
                if vecino in cerrados:
                    print(f"   ⚠️ ¡ALERTA MAYOR! Encontramos una ruta más barata hacia '{vecino}' (Costo nuevo: {nuevo_g}).")
                    print(f"   ⚠️ Como la heurística mintió, tenemos que SACAR a '{vecino}' de los Cerrados y re-evaluarlo.")
                    cerrados.remove(vecino) # ¡Re-trabajo!
                
                g_costos[vecino] = nuevo_g
                f_nuevo = nuevo_g + h[vecino]
                heapq.heappush(frontera, (f_nuevo, vecino))
                print(f"   -> Vemos '{vecino}' desde '{actual}' | g={nuevo_g}, f={f_nuevo}")

a_estrella_tramposo()
```
