```python
def crear_item(nombre, precio, categoria):
    return {
        "nombre": nombre,
        "precio":precio,
        "categoria": categoria
    }
```

```python
inventario = []
inventario.append(crear_item("juancho",32323, "personas"))
print(inventario)
```

```python
def agregar_al_inventario(inventario, nombre, categoria, precio):
    inventario.append(crear_item(nombre,categoria,precio))
    return inventario
```

```python
def buscar_item(inventario, nombre_buscar):
    # funcion para encontrar el nombre, comparacion de strings y accediendo a un elemento del diccionario
    for item in inventario:
        if(item["nombre"].lower() == nombre_buscar.lower()):
            return item

    return "nombre no encontrado"
            
```

```python
def filtrar_categoria(inventario,categoria):
    lista = []
    for item in inventario:
        if item["categoria"].lower() == categoria:
            #list comprehension method
            #return [item  for item in inventario if item["categoria"].lower() == categoria.lower()]
            #normal method
            lista.append(item)
            
    return lista
```
