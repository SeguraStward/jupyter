# EJERCICIO 2 — Aplicación de A* a un problema real simplificado
## Enunciado

Se desea diseñar un sistema de navegación para un robot que se mueve en una cuadrícula 10x10.

El robot:
- Puede moverse arriba, abajo, izquierda, derecha.
- Cada movimiento cuesta 1.
- Algunas celdas están bloqueadas (obstáculos).

Se debe encontrar la ruta más corta desde una posición inicial hasta una meta.

## Parte A — Modelado formal

Defina formalmente:

- Espacio de estados 𝑆
- Función de transición 𝑇
- Función de costo 𝑐(𝑛,𝑛′)
- Estado inicial 𝑠0
- Conjunto de metas 𝐺

## Parte B — Heurística
- Defina la heurística Manhattan.
- Justifique matemáticamente por qué es admisible.
- Demuestre si es consistente.

## Parte C — Implementación

Implemente:

- Greedy Best First Search.
- A*.

Use el mismo mapa.

## Parte D — Análisis comparativo

Compare:
- ¿Cuál encuentra solución más rápido?
- ¿Cuál expande más nodos?
- ¿Cuál garantiza optimalidad?
- ¿Qué ocurre si multiplica Manhattan por 1.5?

## Parte E — Reflexión IA real

Explique:
- ¿Dónde se usa esto en IA real?
- ¿Qué pasaría si el mapa fuera dinámico?
- ¿Cómo afectaría incertidumbre en costos?

## Objetivo conceptual

Que entiendan:
- Aplicación real de heurísticas.
- Diferencia entre Greedy y A*.
- Impacto de sobreestimar.
- Conexión con robótica y videojuegos.
