```python
# ============================================================
# Greedy Best-First Search (GBFS) para el 8-Puzzle (3x3)
# ============================================================
# - Este notebook implementa SOLO Greedy Best-First Search.
# - GBFS elige qué expandir usando únicamente la heurística:
#       f(n) = h(n)
# - NO considera el costo acumulado g(n). (Eso lo veremos en A*)
#
# Representación de estado:
#   Un estado es una tupla de 9 enteros (0..8) en orden fila-major:
#       (a0,a1,a2,
#        a3,a4,a5,
#        a6,a7,a8)
#   Donde 0 representa el espacio vacío.
#
# Ejemplo:
#   (2,8,3,
#    1,6,4,
#    7,0,5)
# ============================================================

import heapq
import time


# -----------------------------
# Utilidad: imprimir el tablero
# -----------------------------
def print_board(state):
    """Imprime un estado del 8-puzzle en forma 3x3."""
    for r in range(3):
        row = state[r * 3:(r + 1) * 3]
        # Mostramos '_' en vez de 0 para que sea más legible
        print(" ".join("_" if x == 0 else str(x) for x in row))
    print()


# -----------------------------------------
# Generación de sucesores (movimientos del 0)
# -----------------------------------------
def get_neighbors(state):
    """
    Devuelve una lista de estados vecinos alcanzables desde 'state'
    moviendo el espacio vacío (0) arriba/abajo/izq/der.
    """
    neighbors = []

    # 1) Encontramos dónde está el 0 (índice 0..8)
    zero_idx = state.index(0)

    # 2) Convertimos el índice a (fila, col)
    #    divmod(i,3) devuelve (i//3, i%3)
    row, col = divmod(zero_idx, 3)

    # 3) Movimientos posibles (delta fila, delta col)
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # up, down, left, right

    for dr, dc in moves:
        new_row, new_col = row + dr, col + dc

        # 4) Verificamos si el movimiento queda dentro del tablero 3x3
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_idx = new_row * 3 + new_col

            # 5) Generamos el nuevo estado intercambiando 0 con la ficha adyacente
            new_state = list(state)
            new_state[zero_idx], new_state[new_idx] = new_state[new_idx], new_state[zero_idx]

            neighbors.append(tuple(new_state))

    return neighbors


# -----------------------------------------
# Heurística: Distancia Manhattan (8-puzzle)
# -----------------------------------------
def manhattan(state, goal):
    """
    Distancia Manhattan para el 8-puzzle:
    Suma, para cada ficha (1..8), la distancia (en cuadritos) entre
    su posición actual y su posición objetivo.

    Fórmula por ficha i:
        |x_i - x_i*| + |y_i - y_i*|
    Heurística total:
        h(n) = sum_{i=1..8} Manhattan(i)
    """
    dist = 0

    # Recorremos las fichas del 1 al 8 (ignoramos el 0)
    for value in range(1, 9):
        cur_idx = state.index(value)
        goal_idx = goal.index(value)

        cur_r, cur_c = divmod(cur_idx, 3)
        goal_r, goal_c = divmod(goal_idx, 3)

        dist += abs(cur_r - goal_r) + abs(cur_c - goal_c)

    return dist


# -----------------------------------------
# Reconstrucción del camino (padres)
# -----------------------------------------
def reconstruct_path(parent, start, goal):
    """
    Reconstruye el camino desde start hasta goal usando el diccionario parent:
        parent[estado] = estado_anterior
    """
    path = []
    node = goal
    while node != start:
        path.append(node)
        node = parent[node]
    path.append(start)
    path.reverse()
    return path


# ============================================================
# Greedy Best-First Search
# ============================================================
def greedy_best_first_search(start, goal, heuristic=manhattan, max_expansions=200000):
    """
    Implementación de Greedy Best-First Search:

    - Estructura principal: 'open_list' como heap (cola de prioridad)
    - Cada elemento del heap:
        (h(estado), contador, estado)
      El 'contador' evita problemas cuando h empata (Python no puede comparar tuplas
      si el tercer elemento fuera el estado directamente y hay empates raros).
    - visited: conjunto de estados ya "cerrados"/procesados

    Retorna:
        path, stats
    donde:
        - path: lista de estados desde start hasta goal (o None si no encuentra)
        - stats: diccionario con métricas (expandidos, tiempo, etc.)
    """
    t0 = time.time()

    # Cola de prioridad: siempre sacamos el estado con menor h(n)
    open_list = []
    push_count = 0  # contador para desempatar

    # Parent para reconstruir el camino cuando encontremos la meta
    parent = {start: None}

    # visited/cerrados: estados ya expandidos
    visited = set()

    # Métricas
    expanded = 0
    generated = 0

    # Insertamos el estado inicial, prioridad = h(start)
    heapq.heappush(open_list, (heuristic(start, goal), push_count, start))
    push_count += 1

    while open_list and expanded < max_expansions:
        # 1) Elegimos el estado con menor heurística (lo más "prometedor")
        h_val, _, current = heapq.heappop(open_list)

        # 2) Si ya lo expandimos antes, lo saltamos
        if current in visited:
            continue

        # 3) Lo marcamos como visitado y lo contamos como expandido
        visited.add(current)
        expanded += 1

        # (Opcional) Si quieres ver el progreso cada cierto número de expansiones
        # if expanded % 5000 == 0:
        #     print(f"[GBFS] Expandidos: {expanded}, h(current)={h_val}")

        # 4) Prueba de meta
        if current == goal:
            t1 = time.time()
            path = reconstruct_path(parent, start, goal)
            stats = {
                "expanded": expanded,
                "generated": generated,
                "visited": len(visited),
                "time_sec": t1 - t0,
                "solution_depth": len(path) - 1,
            }
            return path, stats

        # 5) Expandimos: generamos sucesores
        for nxt in get_neighbors(current):
            generated += 1
            # Importante:
            # En Greedy, la "prioridad" del vecino es SOLO su heurística.
            # No sumamos costos, no usamos g(n).
            if nxt not in visited and nxt not in parent:
                parent[nxt] = current  # guardamos padre solo la primera vez
                heapq.heappush(open_list, (heuristic(nxt, goal), push_count, nxt))
                push_count += 1

    # Si salimos del while, no encontramos solución (o límite de expansiones)
    t1 = time.time()
    stats = {
        "expanded": expanded,
        "generated": generated,
        "visited": len(visited),
        "time_sec": t1 - t0,
        "solution_depth": None,
        "stopped_reason": "open_list_empty" if not open_list else "max_expansions_reached",
    }
    return None, stats


```

```python
initial = (2, 8, 3,
           1, 6, 4,
           7, 0, 5)

goal = (1, 2, 3,
        8, 0, 4,
        7, 6, 5)

print("Estado inicial:")
print_board(initial)

print("Estado meta:")
print_board(goal)

path, stats = greedy_best_first_search(initial, goal, heuristic=manhattan)

print("=== RESULTADO GBFS ===")
print("Stats:", stats)

if path is None:
    print("\nNo se encontró solución (o se alcanzó el límite).")
else:
    print(f"\nSolución encontrada con {len(path)-1} movimientos.\n")
    # Imprime el camino (puede ser largo, lo puedes limitar si quieres)
    for i, st in enumerate(path):
        print(f"Paso {i}:")
        print_board(st)
```
