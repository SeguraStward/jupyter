# Ejercicio de la Clase 4
## Sistema Experto Interactivo de Diagnóstico Académico en Streamlit


Desarrolle una aplicación en Streamlit que modele conocimiento mediante reglas IF–THEN para analizar el perfil académico de un estudiante.

La aplicación debe permitir ingresar datos del usuario mediante una interfaz interactiva y, a partir de estos, inferir uno o más perfiles académicos usando razonamiento basado en reglas.

**El sistema debe:**

- representar conocimiento mediante reglas
- aplicar inferencia automática
- mostrar conclusiones
- explicar el razonamiento seguido
- visualizar resultados de forma clara

**Requisitos obligatorios**
**1. Interfaz**

La solución debe usar al menos 8 componentes de Streamlit, incluyendo:
- panel principal
- barra lateral
- controles de entrada
- botón de análisis
- sección de resultados
- sección de explicación

**2. Base de hechos**

El sistema debe capturar al menos 10 hechos o variables.

Ejemplos:

- horas de estudio
- nivel de estrés
- asistencia
- sueño
- motivación
- organización
- uso del celular
- entrega tardía
- dificultad de concentración
- apoyo recibido

**3. Base de reglas**

Debe incluir al menos 8 reglas.

**4. Conclusiones**

Debe ser capaz de generar al menos 4 perfiles o diagnósticos diferentes.

**5. Explicabilidad**

Debe mostrar:

- hechos detectados
- reglas activadas
- conclusión principal
- justificación

**6. Visualización**

Debe incluir al menos una de estas opciones:

- métricas
- barras de progreso
- ranking de perfiles
- panel de conclusiones
- Requisitos extra

Para quien termine rápido:
- permitir exportar resultados
- mostrar color según nivel de riesgo
- usar st.tabs
- mostrar historial de inferencia
- permitir reiniciar análisis
- añadir recomendaciones específicas por perfil

**7. Componentes Streamlit sugeridos**
- Barra lateral
    - Para entrada general del estudiante:
        - nombre
        - edad
        - carrera
        - semestre

- Cuerpo principal

    - Para hábitos y señales:

       - sliders
       - checkboxes
       - selectbox
       - multiselect

- Resultado

    - Mostrar:

        - perfil principal
        - nivel de riesgo
        - explicación
        - reglas activadas
