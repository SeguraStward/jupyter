# EJERCICIO 1 — Diseño y análisis formal de heurística
## Enunciado

Se desea resolver el problema del 8-puzzle utilizando búsqueda heurística.

Se proponen las siguientes tres heurísticas:

- ℎ1(𝑛) = Número de fichas fuera de lugar.
- ℎ2(𝑛) = Distancia Manhattan.
- ℎ3(𝑛) = 2×Distancia Manhattan.

## Parte A — Implementación
1. Implemente el algoritmo A*.
2. Implemente las tres heurísticas.
3. Ejecute A* desde un estado inicial hasta el estado meta usando cada heurística.

## Parte B — Análisis Experimental

Para cada heurística, reporte:
- Número de nodos expandidos.
- Profundidad de la solución.
- Costo total.
- Tiempo de ejecución.
- Compare los resultados.

## Parte C — Análisis Teórico

Responda y justifique matemáticamente:
- ¿Cuál heurística es admisible?
- ¿Cuál es consistente?
- ¿Qué ocurre con ℎ3(𝑛)?
- ¿Por qué A* con ℎ3(𝑛) puede dejar de ser óptimo?
- ¿Qué relación observa entre calidad de heurística y nodos expandidos?


## Objetivo conceptual

Que comprendan:

- Cómo se construye una heurística.
- Qué significa sobreestimar.
- Cómo afecta la optimalidad.
- Relación teoría ↔ comportamiento empírico.
